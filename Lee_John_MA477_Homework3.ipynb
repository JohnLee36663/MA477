```python
#Enter your code here 

x=np.linspace(-4,4,50)
fig=plt.figure(figsize=(8,6))

axes1=fig.add_axes([0.1,0.1,0.9,0.9])
axes2=fig.add_axes([0.43,0.2,0.25,0.4])

axes1.plot(x,3*np.exp(-0.25*(x**2)),'r-.',lw = 3, label = '3$e^{-0.25x^2}$')
axes1.plot(x,2.8*np.exp(-0.15*((x-0.1)**2)),'b-',lw=0.5,alpha=1,marker='o',markerfacecolor='y',
           markersize=10,markeredgecolor='k',markeredgewidth = 2,label = '2.8$e^{-0.15(x-0.1)^2}$')
axes1.set_title("Many Plots", fontsize=20)

axes2.text(-1.6,2,'Cool',size=15, color = 'r')
axes2.text(-0.4,0,'Plot',size=15, color = 'b')
axes2.text(1.2,-2,'Bro',size=15, color = 'g')
axes2.set_title("Small Plot", fontsize=12)

axes1.legend(loc='upper right')

axes1.set_xlabel('That took a while',fontsize=16)
plt.yticks(np.linspace(-3,3,7))
plt.xticks(np.linspace(-2,2,5))
```

<b> Recreate the following `jointplot` </b>

<font size='4' color='red'>5pts</font>

```python
#Enter code here
sns.set_style('white')
plt.rcParams['xtick.bottom'] = True
plt.rcParams['ytick.left'] = True
ax1 =sns.jointplot(x='number_of_reviews',y='price',data=abnb,height=6,kind='kde')
ax1.ax_joint.set_xticks(np.linspace(0,400,5))
ax1.ax_joint.set_xlabel('number_of_reviews',fontsize = 12)
ax1.ax_joint.set_ylabel('price',fontsize = 12)
plt.show()
```

<b>Time Series: Recreate the following plot</b>

<font color='red' size='4'>10pts</font>

(Hint: Convert the column `last_review` to `DateTime` format and reset it as the index of the dataframe)

```python
#Enter answer here 
abnb['last_review']=pd.to_datetime(abnb['last_review'])
abnb.set_index('last_review')
df1 = pd.melt(abnb,id_vars = ["last_review"],value_vars = ['number_of_reviews','price'])
fig=plt.figure(figsize=(14,6))
sns.set(font_scale = 1,style = 'white')
plt.rcParams['xtick.bottom'] = True
plt.rcParams['ytick.left'] = True


ax=sns.lineplot(x = 'last_review',y='value',hue = 'variable',data = df1)
ax.set_xlabel('last_review',fontsize = 14)
ax.set_ylabel(None,fontsize = 14)
ax.set_xlim('2016-10-01','2019-04-01')
ax.legend(loc = 'upper left',title = None)
ax.set_title('Fluctuations in Number of reviews and Price over time',fontsize = 18)
```
